```python
import pandas as pd
df = pd.read_csv('orders.csv', na_values =['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
df.head(20)
```

```python
#RENAME COLUMNS NAMES...(Lower case and replace space with underscore
#df.raname(columns={'Order Id': 'order_id', 'City':'city', })
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new column discount, sale price and profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']=df['list_price']-df['discount']
df['profit']= df['sale_price']-df['cost_price']
df.head(5)
```

```python
df.dtypes
```

```python
#convert order_date from object data type to datetime
df['order_date']= pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
#drop cost_price, list_price and discount percent columns
df.drop(columns=['cost_price', 'list_price', 'discount_percent'],inplace=True)
```
